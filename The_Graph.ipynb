```python
import matplotlib.pyplot as plt #for plotting purposes
import numpy as np #for computations
from mpl_toolkits.mplot3d import Axes3D #for enabling 3D plots
```

```python
data = np.loadtxt('Datapoints.csv', delimiter = ',') #loading in datapoints csv

#plotting it
plt.figure(figsize = (12,12))
plt.scatter(data[1], data[0], c=data[2])
plt.xlabel('Theta')
plt.ylabel('Phi')
plt.colorbar()

#plt.savefig('10000P_Phi_V_Theta')
```

```python
#for plotting values on heat map around the average distance found in the CSV Reading notebook within some bound
red1 = []
red2 = []
red3 = []
for i in range(len(data[0])):
    if 0.770 < data[2][i] <= 0.798:
        red1.append(data[0][i])
        red2.append(data[1][i])
        red3.append(data[2][i])
```

```python
ded1 = []
ded2 = []
ded3 = []
for i in range(len(data[0])):
    if 0.798 < data[2][i] <= 0.826:
        ded1.append(data[0][i])
        ded2.append(data[1][i])
        ded3.append(data[2][i])
        
plt.figure(figsize = (12,12))
plt.scatter(data[1], data[0], c=data[2])
plt.scatter(ded2, ded1, c='red')
plt.scatter(red2, red1, c='orange')
plt.xlabel('Theta')
plt.ylabel('Phi')
plt.colorbar()
```
